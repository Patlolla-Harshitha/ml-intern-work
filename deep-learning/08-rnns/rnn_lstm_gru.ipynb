# Sequential Data

Sequential data refers to data where the order of observations is important, and each observation depends on previous observations.

## Examples
- Time-series data (stock prices, weather forecasting)
- Speech recognition
- Natural language processing
- Sensor data

## Key Characteristics
- Temporal dependencies exist.
- The position of each element carries meaning.
- Traditional neural networks cannot effectively capture sequence information.

## Interview Question

**Q:** What is sequential data?

**A:** Sequential data is a type of data in which the order of observations matters, and the current observation may depend on previous observations.

# Vanilla Recurrent Neural Network (RNN)

A Recurrent Neural Network is a neural network designed to process sequential data by maintaining a hidden state that stores information from previous time steps.

## Advantages
- Captures short-term dependencies.
- Suitable for sequence modeling tasks.

## Limitations
- Suffers from vanishing gradients.
- Difficult to learn long-term dependencies.

## Applications
- Language modeling
- Speech recognition
- Sequence classification

## Interview Question

**Q:** Why are RNNs suitable for sequential data?

**A:** RNNs maintain hidden states that allow information from previous time steps to influence future predictions.

# Vanishing Gradient Problem

During backpropagation through time, gradients may become extremely small as they propagate backward across many time steps.

## Effects
- Early layers learn very slowly.
- Long-term dependencies are not captured effectively.

## Consequences
- Poor performance on long sequences.
- Difficulty in training traditional RNNs.

## Interview Question

**Q:** What is the vanishing gradient problem?

**A:** It occurs when gradients shrink significantly during backpropagation, preventing the network from learning long-range dependencies.

# Exploding Gradient Problem

Exploding gradients occur when gradients grow excessively during backpropagation.

## Effects
- Unstable learning process.
- Extremely large weight updates.

## Solutions
- Gradient clipping.
- Proper weight initialization.
- Using advanced architectures such as LSTM and GRU.

## Interview Question

**Q:** How can exploding gradients be addressed?

**A:** Gradient clipping is commonly used to limit the magnitude of gradients during training.

# Long Short-Term Memory (LSTM)

LSTM is an advanced type of Recurrent Neural Network designed to overcome the vanishing gradient problem. It introduces a memory cell and gating mechanisms that help preserve information over long sequences.

## Components of LSTM

### 1. Forget Gate
Determines which information from the previous cell state should be discarded.

### 2. Input Gate
Controls which new information should be stored in the cell state.

### 3. Cell State
Acts as the long-term memory of the network.

### 4. Output Gate
Determines what information should be passed to the next hidden state.

## Advantages

- Captures long-term dependencies.
- Reduces the vanishing gradient problem.
- Effective for time-series forecasting and sequence prediction.

## Applications

- Time-series forecasting
- Speech recognition
- Machine translation
- Text generation

## Interview Question

**Q:** Why is LSTM preferred over Vanilla RNN?

**A:** LSTM uses gating mechanisms and memory cells that allow it to learn long-term dependencies more effectively, overcoming the vanishing gradient problem.

# Gated Recurrent Unit (GRU)

GRU is a simplified version of LSTM that combines certain gates to reduce computational complexity while maintaining strong performance.

## Components of GRU

### 1. Update Gate
Determines how much past information should be retained.

### 2. Reset Gate
Determines how much past information should be forgotten.

## Advantages

- Faster training compared to LSTM.
- Requires fewer parameters.
- Performs well on many sequence modeling tasks.

## Difference Between LSTM and GRU

| Feature | LSTM | GRU |
|----------|------|-----|
| Gates | 3 | 2 |
| Cell State | Present | Absent |
| Parameters | More | Fewer |
| Training Speed | Slower | Faster |

## Interview Question

**Q:** What is the main difference between LSTM and GRU?

**A:** LSTM has separate memory cells and three gates, whereas GRU uses two gates without an explicit cell state, making it computationally simpler.

```python
# Extract passenger values

data = df[['#Passengers']].values

data[:5]
```

```python
# Feature Scaling

scaler = MinMaxScaler(
    feature_range=(0, 1)
)

scaled_data = scaler.fit_transform(
    data
)

print("Scaled Data Shape:",
      scaled_data.shape)

scaled_data[:5]
```

# Sequence Creation

LSTM models require sequential input data.

A sliding window approach is used to create input-output pairs.

For example, if the sequence length is 12 months:

Input:
Jan → Dec

Output:
Next month's passenger count

```python
# Function to create sequences

def create_sequences(
        dataset,
        sequence_length):
    
    X = []
    y = []

    for i in range(
            len(dataset)
            - sequence_length):
        
        X.append(
            dataset[
                i:i+sequence_length,
                0
            ]
        )

        y.append(
            dataset[
                i+sequence_length,
                0
            ]
        )

    return (
        np.array(X),
        np.array(y)
    )
```

```python
# Create sequences

sequence_length = 12

X, y = create_sequences(
    scaled_data,
    sequence_length
)

print("X Shape:",
      X.shape)

print("y Shape:",
      y.shape)
```

```python
# Train-Test Split

train_size = int(
    len(X) * 0.8
)

X_train = X[:train_size]
X_test = X[train_size:]

y_train = y[:train_size]
y_test = y[train_size:]

print("Training Samples:",
      len(X_train))

print("Testing Samples:",
      len(X_test))
```

```python
# Reshape for LSTM

X_train = np.reshape(
    X_train,
    (
        X_train.shape[0],
        X_train.shape[1],
        1
    )
)

X_test = np.reshape(
    X_test,
    (
        X_test.shape[0],
        X_test.shape[1],
        1
    )
)

print("X_train Shape:",
      X_train.shape)

print("X_test Shape:",
      X_test.shape)
```

```python
# Build LSTM Model

model = Sequential([
    LSTM(
        units=50,
        activation='tanh',
        input_shape=(
            X_train.shape[1],
            1
        )
    ),
    Dense(1)
])

model.summary()
```

```python
# Compile the Model

model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)
```

```python
# Train the Model

history = model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=16,
    validation_data=(
        X_test,
        y_test
    ),
    verbose=1
)
```

```python
# Plot Training and Validation Loss

plt.figure(figsize=(10, 5))

plt.plot(
    history.history['loss'],
    label='Training Loss'
)

plt.plot(
    history.history['val_loss'],
    label='Validation Loss'
)

plt.title(
    'Training vs Validation Loss'
)

plt.xlabel('Epoch')

plt.ylabel('Loss')

plt.legend()

plt.grid(True)

plt.show()
```

```python
# Generate Predictions

train_predictions = model.predict(
    X_train
)

test_predictions = model.predict(
    X_test
)
```

```python
# Inverse Transform Predictions

train_predictions = scaler.inverse_transform(
    train_predictions
)

test_predictions = scaler.inverse_transform(
    test_predictions
)

y_train_actual = scaler.inverse_transform(
    y_train.reshape(-1, 1)
)

y_test_actual = scaler.inverse_transform(
    y_test.reshape(-1, 1)
)
```

```python
# Calculate Evaluation Metrics

mae = mean_absolute_error(
    y_test_actual,
    test_predictions
)

rmse = np.sqrt(
    mean_squared_error(
        y_test_actual,
        test_predictions
    )
)

print(
    "Mean Absolute Error:",
    round(mae, 2)
)

print(
    "Root Mean Squared Error:",
    round(rmse, 2)
)
```

```python
# Visualize Actual vs Predicted Values

plt.figure(figsize=(12, 6))

plt.plot(
    y_test_actual,
    label='Actual Passengers'
)

plt.plot(
    test_predictions,
    label='Predicted Passengers'
)

plt.title(
    'Actual vs Predicted Passenger Counts'
)

plt.xlabel('Time')

plt.ylabel('Number of Passengers')

plt.legend()

plt.grid(True)

plt.show()
```
